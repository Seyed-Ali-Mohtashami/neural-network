```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array(range(-5,6))
```

```python
# print('1.Relu : With default values, this returns the standard ReLU activation: max(x, 0), the element-wise maximum of 0 and the input tensor.')
layer=  tf.keras.layers.Activation(tf.nn.relu)
y_relu=layer(x)

plt.figure(figsize=(8,4))
plt.subplot(121)
plt.xticks([-10,0,+10])
plt.yticks([-10,0,+10])
plt.title('x')
plt.plot(x,x)
plt.subplot(122)
plt.title('Relu')
plt.xticks([])
plt.yticks([])
plt.plot(x,x)
plt.plot(x,y_relu)
plt.show()
```

```python
y_exponential=tf.keras.activations.exponential(np.float32(x))

plt.figure(figsize=(8,4))

plt.subplot(121)
plt.xticks([-10,0,+10])
plt.yticks([-10,0,+10])
plt.title('x')
plt.plot(x,x)

plt.subplot(122)
plt.title('exponential')
plt.xticks([-10,0,+10])
plt.yticks([-10,0,+10])
plt.plot(x,x)
plt.plot(x,y_exponential)

plt.show()
```

```python
#linear
plt.figure(figsize=(12,4))

def f(x):
  fx=np.zeros(len(x))
  for i in range(len(x)):
     fx[i]= x[i]
  return fx

x=np.arange(-10,11)

# print(x)
plt.subplot(131)
plt.plot(x,x )
plt.title('x')

plt.subplot(132)
# print(f(x))
plt.plot(x, f(x))
plt.xticks([-10,0,10])
plt.yticks([-10,0,10])
plt.title('linear')

dfdx=np.gradient(f(x),x)
# print(x)
# print(f(x))
# print(dfdx)
plt.subplot(133)
plt.plot(x, dfdx)
plt.title('gradient')
plt.xticks([-10,0,10])
plt.yticks([-10,0,10])

plt.show()
```

```python
#sigmiod
plt.figure(figsize=(12,4))

def f(x):
  fx=np.zeros(len(x))
  for i in range(len(x)):
     fx[i]= 1/(1 + np.exp(-x[i]))
  return fx

x=np.arange(-10,11)

# print(x)
plt.subplot(131)
plt.plot(x,x )
plt.title('x')

plt.subplot(132)
# print(f(x))
plt.plot(x, f(x))
plt.xticks([-10,0,10])
plt.yticks([0,0.5])
plt.title('sigmoid')

dfdx=np.gradient(f(x),x)
# print(x)
# print(f(x))
# print(dfdx)
plt.subplot(133)
plt.plot(x, dfdx)
plt.title('gradient')
plt.xticks([-10,0,10])
plt.yticks([-0.5,0,0.5])

plt.show()
```

```python
#hyperbolic tangent
plt.figure(figsize=(12,4))

def f(x):
  fx=np.zeros(len(x))
  for i in range(len(x)):
     fx[i]= (np.exp(x[i])-np.exp(-x[i]))/(np.exp(x[i]) + np.exp(-x[i]))
  return fx

x=np.arange(-10,11)

# print(x)
plt.subplot(131)
plt.plot(x,x )
plt.title('x')

plt.subplot(132)
# print(f(x))
plt.plot(x, f(x))
plt.xticks([-10,0,10])
plt.yticks([-1,0,1])
plt.title('tanh')

dfdx=np.gradient(f(x),x)
# print(x)
# print(f(x))
# print(dfdx)
plt.subplot(133)
plt.plot(x, dfdx)
plt.title('gradient')
plt.xticks([-10,-2,0,2,10])
plt.yticks([-0.5,0,0.5])

plt.show()
```

```python
#relU
plt.figure(figsize=(12,4))

def f(x):
  fx=np.zeros(len(x))
  for i in range(len(x)):
     fx[i]= max(x[i],0)
  return fx

x=np.arange(-10,11)

# print(x)
plt.subplot(131)
plt.plot(x,x )
plt.title('x')

plt.subplot(132)
# print(f(x))
plt.plot(x, f(x))
plt.xticks([-10,0,10])
plt.yticks([-10,0,10])
plt.title('relu')

dfdx=np.gradient(f(x),x)
# print(x)
# print(f(x))
# print(dfdx)
plt.subplot(133)
plt.plot(x, dfdx)
plt.title('gradient')
plt.xticks([-10,0,10])
plt.yticks([-10,0,10])

plt.show()
```

```python
#ElU
plt.figure(figsize=(12,4))

def f(x):
  fx=np.zeros(len(x))
  for i in range(len(x)):
    if x[i] >= 0:
      fx[i]=x[i]
    else:
      fx[i]= np.exp(x[i])- 1
  return fx

x=np.arange(-10,11)

# print(x)
plt.subplot(131)
plt.plot(x,x )
plt.title('x')

plt.subplot(132)
# print(f(x))
plt.plot(x, f(x))
plt.xticks([-10,0,10])
plt.yticks([-10,0,10])
plt.title('elu')

dfdx=np.gradient(f(x),x)
# print(x)
# print(f(x))
# print(dfdx)
plt.subplot(133)
plt.plot(x, dfdx)
plt.title('gradient')
plt.xticks([-10,0,10])
plt.yticks([-10,0,10])

plt.show()
```

```python
#leaky relu
plt.figure(figsize=(12,4))

def f(x):
  fx=np.zeros(len(x))
  for i in range(len(x)):
    fx[i]= max(0.2*x[i] , x[i])
  return fx

x=np.arange(-10,11)

# print(x)
plt.subplot(131)
plt.plot(x,x )
plt.title('x')

plt.subplot(132)
# print(f(x))
plt.plot(x, f(x))
plt.xticks([-10,0,10])
plt.yticks([-10,0,10])
plt.title('Leakyrelu')

dfdx=np.gradient(f(x),x)
# print(x)
# print(f(x))
# print(dfdx)
plt.subplot(133)
plt.plot(x, dfdx)
plt.title('gradient')
plt.xticks([-10,0,10])
plt.yticks([-10,0.2,10])

plt.show()
```
